```python
import os
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
actors = ['dwayne_johnson', 'chris_hemsworth', 'robert_downey_jr', 'akshay_kumar',
          'jackie_chan', 'bradley_cooper', 'adam_sandler',
          'scarlett_johansson', 'sofia_vergara', 'chris_evans']

for a in actors:
    with open(a + '.txt', 'r') as f:
        text = f.readline()
        text_tokens = word_tokenize(text)
        tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]
        tokens_without_sw = ' '.join(tokens_without_sw)
        wordcloud = WordCloud().generate(tokens_without_sw)

    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.savefig(a + '.png')
```
