```python
import requests
from bs4 import BeautifulSoup
```

```python
title_page = requests.get('https://www.imdb.com/title/tt2382320/fullcredits/')
title_page_bs = BeautifulSoup(title_page.text)
```

```python
def get_actors_by_movie_soup(cast_page_soup, num_of_actors_limit=None):
    tb = cast_page_soup.find('table', {'class': 'cast_list'})
    actors = tb.find_all('tr', {'class': ['odd', 'even']})
    name_to_link = {}
    for i, a in enumerate(actors):
        if num_of_actors_limit and i >= num_of_actors_limit:
            break

        info = a.find_all('a')[1]
        actor_name = info.text.strip()
        link = info['href']
        name_to_link[actor_name] = link

    return [(k,v) for k, v in name_to_link.items()]

get_actors_by_movie_soup(title_page_bs, 10)
```

```python
headers = {
    'Accept-Language': 'en',
    'X-FORWARDED-FOR': '2.21.184.0'
}
```

```python


actor_page = requests.get('https://www.imdb.com/name/nm0397171/', headers=headers)
actor_page_soup = BeautifulSoup(actor_page.text)
```

```python
def get_movies_by_actor_soup(actor_page_soup, num_of_movies_limit=None):
    node = actor_page_soup.find('div', id='filmo-head-actor')
    if not node:
        node = actor_page_soup.find('div', id='filmo-head-actress')

    tb = node.findNext('div')
    filmos = tb.select('div.filmo-row')
    name_to_link = {}
    counter = 0
    for f in filmos:
        if num_of_movies_limit and counter >= num_of_movies_limit:
            break

        info = f.find_all('a')
        matches = ['(TV Series)', '(TV Special)', '(announced)', '(pre-production)', '(post-production)', '(completed)', '(Video Game)', '(Short) ', '(TV Mini Series)', '(Video short)', '(Video)', '(TV Movie)', '(Music Video short)']
        if any(x in f.text for x in matches):
            continue

        title = info[0]
        name = title.text
        link = title['href']
        name_to_link[name] = link
        counter += 1

    return [(k,v) for k, v in name_to_link.items()]

get_movies_by_actor_soup(actor_page_soup)
```
